### Regardless of the method of initialization, HCSSMF can always reach a unique solution (the optimal solution).

For the sake of convenience, we first initialize with 0 (for a regular symmetric matrix decomposition, 0 is one of its stationary points;  initializing with 0 will not yield results), and fix the random seed for the sampling process at 42 (ensuring consistent data labeling for each run) for experimentation.

```python
DataName = 'COIL20'
run(DataName, rate, Mtime = Mtime, random_seed = 42)
```

```python
DataName = 'COIL100_obj'
run(DataName, rate, Mtime = Mtime, random_seed = 42)
```

When both the data and initialization are fixed, HCSSMF will produce the same values every time it runs, which is predictable. Next, we randomly initialize Z and output the first 5 numbers of Z to visually demonstrate that their initializations are different each time (note that our random initialization includes negative values, which is typically prohibited for traditional NMF but does not affect our algorithm).

```python
DataName = 'COIL20'
run(DataName, rate, Mtime = Mtime, random_seed = 42, random_init = True)
```

```python
DataName = 'COIL100_obj'
run(DataName, rate, Mtime = Mtime, random_seed = 42, random_init = True)
```

Even when initialized randomly, our algorithm consistently produces the same results (reaching the global optimum) in every run.
